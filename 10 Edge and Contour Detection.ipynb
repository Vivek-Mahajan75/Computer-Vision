```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('lenna.png',0)
```

```python
image.shape
```

```python
# Sobel parameters : ksize,dx,dy,depth
sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize = 3)
sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize = 3)

# Combine the sobwl images(optional)
combined_sobel = cv2.add(sobelx,sobely)
```

```python
thresh,binary_sobel = cv2.threshold(combined_sobel,100,255,cv2.THRESH_BINARY)
# adjust threshold value as needed
```

```python
#display results
plt.figure(figsize=(16,8))
plt.subplot(1,4,1)
plt.title('Original')
plt.imshow(image, cmap = 'gray')
plt.subplot(1,4,2)
plt.title('Sobel X')
plt.imshow(sobelx, cmap = 'gray')
plt.subplot(1,4,3)
plt.title('Sobel Y')
plt.imshow(sobely, cmap = 'gray')
plt.subplot(1,4,4)
plt.title('Final')
plt.imshow(combined_sobel, cmap = 'gray')
```

## Scharr Edge detection

```python
# X gradient Scharr operator
fieldx = cv2.Scharr(image,cv2.CV_32F,1,0) / 10.34

# Y gradient Scharr operator
fieldy = cv2.Scharr(image,cv2.CV_32F,1,0) / 10.34

#display results
plt.figure(figsize=(16,8))
plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(image, cmap = 'gray')
plt.subplot(1,3,2)
plt.title('X Value')
plt.imshow(fieldx, cmap = 'gray')
plt.subplot(1,3,3)
plt.title('Y Value')
plt.imshow(fieldy, cmap = 'gray')
```

```python
cv2.imshow('X Image', fieldx)
cv2.imshow('Y Image', fieldy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('X Image', sobelx)
cv2.imshow('Y Image', sobely)
cv2.imshow('Combined', combined_sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Countour Detection

```python
import cv2

img = cv2.imread('rgy.jpg')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #convert to gray if needed

# creates a binary image where pixels above 127 are set to 255
#and other to 0
thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)[1]

# cv2.findCoutours finds contours in the binary image
# cv2.RETR_TREE retrieves all contours and creates a hierarchy tree
# cv2.CHAIN_APPROX_SIMPLE approximates contours to save memory

contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,
                                      cv2.CHAIN_APPROX_SIMPLE)

# -1 draws all contours
# (0,255,0) is the color(green)
# 2 is the thickness of the lines.

cv2.drawContours(img,contours,-1,(0,255,0),2)

cv2.imshow('Original Image', img)
cv2.imshow('Threshold Detection', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
