# Binary thresholding

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('lenna.png')

img = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)  #convert to gray if needed

#threshold to create binary image  
ret , thresh1 = cv2.threshold(img,120,255,cv2.THRESH_BINARY)
ret , thresh2 = cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV)
ret , thresh3 = cv2.threshold(img,120,255,cv2.THRESH_TRUNC)
ret , thresh4 = cv2.threshold(img,120,255,cv2.THRESH_TOZERO)
ret , thresh5 = cv2.threshold(img,120,255,cv2.THRESH_TOZERO_INV)

#display results
plt.figure(figsize=(16,8))
plt.subplot(1,5,1)
plt.title('binary threshold Image')
plt.imshow(thresh1, cmap = 'gray')
plt.subplot(1,5,2)
plt.title('Inverse binary threshold Image')
plt.imshow(thresh2, cmap = 'gray')
plt.subplot(1,5,3)
plt.title('THRESH TRUNC')
plt.imshow(thresh3, cmap = 'gray')
plt.subplot(1,5,4)
plt.title('THRESH TOZERO')
plt.imshow(thresh4, cmap = 'gray')
plt.subplot(1,5,5)
plt.title('THRESH_TOZERO_INV')
plt.imshow(thresh5, cmap = 'gray')
```

## Otsu Thresholding

```python
import cv2
import numpy as np

img = cv2.imread('lenna.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #convert to gray if needed

#threshold to create binary image  
_, thresh_otsu = cv2.threshold(gray,0,255,cv2.THRESH_OTSU)

#display results
cv2.imshow('ORIGINAL IMAGE', img)
cv2.imshow('Otsu Thresholding', thresh_otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Otsu Thresholding using skimage

```python
from skimage import io,filters
image = io.imread('lenna.png',as_gray=True)
```

```python
threshold = filters.threshold_otsu(image)
binary_image = image > threshold
```

```python
import matplotlib.pyplot as plt

```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize = (10,5))

ax1.imshow(image,cmap = 'gray')
ax1.set_title("Original grayscale image")

ax2.imshow(binary_image,cmap = 'gray')
ax2.set_title("Binary image after Otsu's thresholding")

plt.show()
```
